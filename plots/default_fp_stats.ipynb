```python
import numpy as np
import healpy as hp
from lsst.sims.featureScheduler.utils import standard_goals
%matplotlib inline
```

```python
sg = standard_goals()
```

```python
hp.mollview(sg['r'])
```

```python
np.unique(sg['r'])
```

```python
tot = sg['r']*0
for key in sg:
    tot += sg[key]
```

```python
wfd = np.where(sg['r'] == 1)
nes = np.where(sg['r'] == 0.46)
gpscp = np.where(sg['r'] == 0.15)
```

```python
hp.mollview(tot)
```

```python
tot_hp = np.sum(tot)
print('wfd=', tot[wfd].sum()/tot_hp)
print('nes=', tot[nes].sum()/tot_hp)
print('gpscp=', tot[nes].sum()/tot_hp)
```

```python
# If we take off 5% for DDF
print('wfd=', tot[wfd].sum()/tot_hp*.95)
print('nes=', tot[nes].sum()/tot_hp*.95)
print('gpscp=', tot[nes].sum()/tot_hp*.95)
```
